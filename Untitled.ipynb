```python
from keras.layers.core import Dense, Activation, Dropout
from keras.layers.recurrent import LSTM
from keras.models import Sequential
import time
import pandas as pd
import matplotlib
%matplotlib notebook
```

```python
from pandas import DataFrame
from pandas import concat
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
import matplotlib.pyplot as plt
# create sequence
length = 10
sequence = [i/float(length) for i in range(length)]
# create X/y pairs
df = DataFrame(sequence)
df = concat([df, df.shift(1)], axis=1)
df.dropna(inplace=True)
# convert to LSTM friendly format
values = df.values
X, y = values[:, 0], values[:, 1]
X = X.reshape(len(X), 1, 1)
# configure network
n_batch = 1
n_epoch = 1000
n_neurons = 10
# design network
model = Sequential()
model.add(LSTM(n_neurons, batch_input_shape=(n_batch, X.shape[1], X.shape[2]), stateful=True))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam',metrics=['accuracy'])
# fit network
loss = []
acc = []
for i in range(n_epoch):
    print '{}/{} epoch'.format(i,n_epoch)
    hist = model.fit(X, y, epochs=1, batch_size=n_batch, verbose=1, shuffle=False)
    loss.append(hist.history['loss'])
    acc.append(hist.history['acc'])
    model.reset_states()
plt.figure(0)
plt.plot(range(n_epoch),loss,'b-')
plt.figure(1)
plt.plot(range(n_epoch),acc,'g-')
plt.show()
```

```python
X_reshaped = X[:,0].reshape(len(X), 1, 1)
scores = model.predict(X_reshaped,batch_size=1)
print("Accuracy: %.2f%%" % (scores[1]*100))
# online forecast
for i in range(len(X)):
    testX, testy = X[i], y[i]
    testX = testX.reshape(1, 1, 1)
    yhat = model.predict(testX, batch_size=1)
    print('>Expected=%.1f, Predicted=%.1f' % (testy, yhat))
```
